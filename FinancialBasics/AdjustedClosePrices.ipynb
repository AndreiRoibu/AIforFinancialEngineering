# Exploring Close and Adjusted Close Prices

```python
df = pd.read_csv('sp500sub.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
goog = df[df['Name']=='GOOG']
goog[['Close', 'Adj Close']].plot(figsize=(12, 6))
```

In this dataset, non-adjusted Close already accounts for stock splits, but not divident payments

```python
aapl = df[df['Name']=='AAPL']
aapl[['Close', 'Adj Close']].plot(figsize=(12, 6))
```

For Apple (AAPL), we can see that there is a difference between the two prices, induced by the payment of dividents
