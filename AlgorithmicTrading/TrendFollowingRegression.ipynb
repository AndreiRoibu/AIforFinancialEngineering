```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('sp500_closefull.csv', index_col=0)
df.head()
```

```python
# clear out the NaNs
df = df.dropna(axis=0, how='all')
df = df.dropna(axis=1, how='any')
print(df.isna().sum().sum())
```

```python
# create a returns dataframe
df_returns = pd.DataFrame()
for col in df.columns:
    df_returns[col] = np.log(df[col]).diff()

# shift SNP500 returns by 1 day
df_returns['SPY'] = df_returns['SPY'].shift(-1)
```

```python
df_returns.head()
```

```python
print(len(df_returns))
```

```python
Ntest = 1000
train = df_returns.iloc[1:-Ntest]
test = df_returns.iloc[-Ntest:-1]
x_cols = ['AAPL', 'MSFT', 'AMZN', 'JNJ', 'V', 'PG', 'JPM'] # select some large cap stocks

Xtrain = train[x_cols]
Ytrain = train['SPY']
Xtest = test[x_cols]
Ytest = test['SPY']
```

```python
Xtrain.head()
```

```python
Ytrain.head()
```

# Linear Regression Model

```python
model = LinearRegression()
model.fit(Xtrain, Ytrain)
print('Train score:', model.score(Xtrain, Ytrain))
print('Test score:', model.score(Xtest, Ytest)) # perfect gives 1.0 and 0.0 gives random
```

```python
# We do not care about the value, but the sign of the prediction (or direction)

Ptrain = model.predict(Xtrain)
Ptest = model.predict(Xtest)

print('Train accuracy:', np.mean(np.sign(Ptrain) == np.sign(Ytrain)))
print('Test accuracy:', np.mean(np.sign(Ptest) == np.sign(Ytest)))
```

```python
# Calculate the train and test returns for the algorithm and for the buy-and-hold strategy
print('Algo Train returns:', np.sum(Ytrain * np.array(Ptrain > 0)))
print('BAH Train returns:', np.sum(Ytrain))
print('---')
print('Algo Test returns:', np.sum(Ytest * np.array(Ptest > 0)))
print('BAH Test returns:', np.sum(Ytest))
```

# Logistic Regression

```python
model = LogisticRegression(C=10) # C is the inverse of regularization strength ie. a way to control overfitting by penalizing large weights
Ctrain = (Ytrain > 0).astype(np.int32)
Ctest = (Ytest > 0).astype(np.int32)
model.fit(Xtrain, Ctrain)
print('Train score:', model.score(Xtrain, Ctrain))
print('Test score:', model.score(Xtest, Ctest))
```

```python
# We do not care about the value, but the sign of the prediction (or direction)

Ptrain = model.predict(Xtrain)
Ptest = model.predict(Xtest)

print(set(Ptrain), set(Ptest))
print('Where are we holding?', np.where(Ptest==0)[0])
```

```python
# Calculate the train and test returns for the algorithm and for the buy-and-hold strategy
print('Algo Train returns:', np.sum(Ytrain * np.array(Ptrain > 0)))
print('BAH Train returns:', np.sum(Ytrain))
print('---')
print('Algo Test returns:', np.sum(Ytest * np.array(Ptest > 0)))
print('BAH Test returns:', np.sum(Ytest))
```
