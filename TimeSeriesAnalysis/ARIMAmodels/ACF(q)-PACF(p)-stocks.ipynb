```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = pd.read_csv('sp500sub.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
goog = df[df['Name']=='GOOG'][['Close']].copy()
aapl = df[df['Name']=='AAPL'][['Close']].copy()
ibm = df[df['Name']=='IBM'][['Close']].copy()
sbux = df[df['Name']=='SBUX'][['Close']].copy()
```

```python
def log_ret(price):
    # Taking the log return to ensure stationarity
    # log return is the difference between the log of the price at time t and the log of the price at time t-1
    # This difference yields a stationary time series
    log_price = np.log(price)
    return log_price.diff()
```

```python
goog['LogRet'] = log_ret(goog['Close'])
aapl['LogRet'] = log_ret(aapl['Close'])
ibm['LogRet'] = log_ret(ibm['Close'])
sbux['LogRet'] = log_ret(sbux['Close'])
```

```python
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(goog['LogRet'].dropna(), ax=ax[0]);
plot_pacf(goog['LogRet'].dropna(), ax=ax[1]);
```

```python
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(aapl['LogRet'].dropna(), ax=ax[0]);
plot_pacf(aapl['LogRet'].dropna(), ax=ax[1]);
```

```python
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(ibm['LogRet'].dropna(), ax=ax[0]);
plot_pacf(ibm['LogRet'].dropna(), ax=ax[1]);
```

```python
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(sbux['LogRet'].dropna(), ax=ax[0]);
plot_pacf(sbux['LogRet'].dropna(), ax=ax[1]);
```
