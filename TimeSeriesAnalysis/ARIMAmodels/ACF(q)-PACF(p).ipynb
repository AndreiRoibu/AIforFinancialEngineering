# Import data and packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

# Random iiD Normal Data

```python
x0 = np.random.randn(1000)
plt.plot(x0)
```

```python
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(x0, ax=ax[0]);
plot_pacf(x0, ax=ax[1]);
```

# Create a list with an AR(1) function

```python
x1 = [0]
for i in range(1000):
    x = 0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)

plt.plot(x1);
```

```python
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(x1, ax=ax[0]);
plot_pacf(x1, ax=ax[1]);
```

# Create an AR(1) process again

```python
x1 = [0]
for i in range(1000):
    x = -0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)

plt.plot(x1);

fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(x1, ax=ax[0]);
plot_pacf(x1, ax=ax[1]);
```

# Create an AR(2) process

```python
x1 = [0, 0]
for i in range(1000):
    x = 0.5 * x1[-1] -0.3 * x1[-2] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)

plt.plot(x1);

fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(x1, ax=ax[0]);
plot_pacf(x1, ax=ax[1]);
```

# Create an AR(5)

```python
x1 = [0, 0, 0, 0, 0]
for i in range(1000):
    x = 0.5 * x1[-1] -0.3 * x1[-2] - 0.6 * x1[-5] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)

plt.plot(x1);

fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(x1, ax=ax[0]);
plot_pacf(x1, ax=ax[1]);
```

# Add iiD Noise

```python
errors = np.random.randn(1000)
plt.plot(errors);
fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(errors, ax=ax[0]);
plot_pacf(errors, ax=ax[1]);
```

# Generate an MA(1) process

```python
errors = 0.1 * np.random.randn(1000)
ma1 = []
for i in range(1000):
    if i >= 1:
        x = 0.5 * errors[i-1] + errors[i]
    else:
        x = errors[i]
    ma1.append(x)
ma1 = np.array(ma1)

plt.plot(ma1);

fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(ma1, ax=ax[0]);
plot_pacf(ma1, ax=ax[1]);
```

# Create MA(2) process

```python
errors = 0.1 * np.random.randn(1000)
ma1 = []
for i in range(1000):
    if i >= 1:
        x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + errors[i]
    else:
        x = errors[i]
    ma1.append(x)
ma1 = np.array(ma1)

plt.plot(ma1);

fig, ax = plt.subplots(2,1, figsize=(12, 8))
plot_acf(ma1, ax=ax[0]);
plot_pacf(ma1, ax=ax[1]);
```
